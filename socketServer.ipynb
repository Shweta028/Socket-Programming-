```python
#!usr/bin/python

#On receiver side

import socket
import matplotlib.pyplot as plt

# giving IP version, type of Protocol

s=socket.socket(socket.AF_INET,socket.SOCK_DGRAM)

s.bind(("192.168.43.19",5000))
print "Displaying bar graph"

x=[]
y=[]


while True:
    data=s.recvfrom(1024)
    if len(data[0])==0:
        print x
        print y

        plt.bar(x,y)
        plt.show()
        continue

    row=(data[0].split(','))
    x.append(int(row[0]))
    y.append(int(row[1]))
    

   

```
